```python
data = pd.read_csv('countries.csv')
```

```python
#compare data of 2 different countries
Albania = data[data.country == 'Albania']
Albania
```

```python
Zimbabwe = data[data.country == 'Zimbabwe']
Zimbabwe 
```

```python
Vietnam = data[data.country == 'Vietnam']
Vietnam
```

```python
plt.plot(Albania.year, Albania.population/10**6, color = "black")
plt.plot(Zimbabwe.year, Zimbabwe.population/10**6, color = "green")
plt.plot(Vietnam.year, Vietnam.population/10**6, color = "red")
plt.title("Production rates in 3 countries")
plt.legend(["Albania", "Zimbabwe", "Vietnam"])
plt.xlabel("year")
plt.ylabel("Average rates of production")
plt.show()
```

```python
data1 = pd.read_csv('covid-vaccination-doses-per-capita.csv')
Africa = data1[data1.Entity == 'Africa']
Africa
data2 = pd.read_csv('covid-vaccination-doses-per-capita.csv')
Bolivia = data1[data1.Entity == "Bolivia"]
Bolivia
```

```python
data3 = pd.read_csv('covid-vaccination-doses-per-capita.csv')
Canada = data3[data3.Entity == "Canada"]
Canada
```

```python
data4 = pd.read_csv('covid-vaccination-doses-per-capita.csv')
China = data3[data4.Entity == "China"]
China
```

```python
plt.plot(Africa.Entity, Africa.Day, Africa.total_vaccinations_per_hundred, color = "black")
plt.plot(Bolivia.Entity, Bolivia.Day, Bolivia.total_vaccinations_per_hundred, color = "green")
plt.plot(Canada.Entity, Canada.Day, Canada.total_vaccinations_per_hundred, color = "red")
plt.plot(China.Entity, China.Day, China.total_vaccinations_per_hundred, color = "yellow")

plt.title("Vaccination rates in 4 countries")
plt.legend(["Africa", "Bolivia", "Canada", "China"])
plt.xlabel("year")
plt.ylabel("Average rise of number of vaccinated people")
plt.show()
```
